```python
import os
import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
import matplotlib.cm as cm

from pyAudioAnalysis import ShortTermFeatures as aSF
from pyAudioAnalysis import MidTermFeatures as aMF
from pyAudioAnalysis import audioBasicIO as aIO

from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import Sequential
from sklearn import metrics

```

## Feature Extraction from Audio Files
---
In this section, we outline a process for extracting features from audio files stored in a directory. The process involves reading each audio file, extracting mid-term features, and compiling these features into a comprehensive DataFrame for analysis.

#### Setting Up the Environment
Initially, we define the directory containing the audio files and prepare containers for the extracted features

#### Iterating Through Audio Files
We iterate over each file in the specified directory, performing several key operations for feature extraction:

1. Skip System Files: Ignore files that are not relevant to the analysis (e.g., system files starting with "._").
2. Read the Audio File: Utilize audio processing libraries to load the audio files.
3. Define Window Steps: Set the parameters for mid-term and short-term feature extraction based on the audio sampling rate.
4. Extract Features: Use the audio processing library to extract mid-term features from each audio file.
5. Compile Features into DataFrame: Construct a DataFrame from the extracted features and add it to a list for later aggregation.

#### Aggregating Extracted Features
After processing all files, we aggregate the individual DataFrames into a single DataFrame for comprehensive analysis:

This aggregated DataFrame, extracted_df, now contains the mid-term features extracted from each audio file, making it ready for further analysis or machine learning applications.

```python
_dir = './ccdata'
file_df_list = []
extracted_df = pd.DataFrame()

file_id = 0
for root, subdirs, files in os.walk(_dir):
    for file in files:
        try:
            # Hard-code fix
            if file.startswith("._"):
                file = file[2:]
            
            path = os.path.join(root, file)
            # print(path)
            fs, s = aIO.read_audio_file(path)
            
            # extract mid-term features using a 50msec non-overlapping windows
            mid_window_step = fs*0.50 # sampling rate * time (sec)
            short_window_step = fs*0.05 # sampling rate * time (sec)
            mt, st, mt_n = aMF.mid_feature_extraction(s, fs,
                                               mid_window_step, mid_window_step,
                                               short_window_step, short_window_step)
            # Create DF of extracted features from file
            row_dict = { "file_id": file_id, "file": file, **{key: mt[j] for j, key in enumerate(mt_n)} }
            file_df_list.append(pd.DataFrame(row_dict))
            # Running counter across directories
            file_id += 1
    
            # print(f'{mt.shape[1]} frames, {mt.shape[0]} mid-term features')
        except:
            print(f"filename '{path}' does not conform to file format standards")

extracted_df = pd.concat(file_df_list, ignore_index=True)
```

```python
# aggregate windows; each file gets a row of averaged data
avg_extracted_df = extracted_df.groupby(['file']).mean()
avg_extracted_df.drop(columns=['file_id'], inplace=True)
avg_extracted_df
```

### Loading and Preparing Labeled Data
In this section, we discuss the procedure to load labeled data from a CSV file and prepare it for further analysis by selecting specific features and renaming columns for consistency.

#### Loading Labeled Data
The first step involves loading the data from a CSV file. This CSV file, named final_classification_features.csv, contains various features extracted from audio files along with their corresponding labels.

#### Selecting Relevant Features
After loading the data, we select a subset of features that are relevant for our analysis. This includes features related to pitch, spectral characteristics, and perceptual attributes, among others. The selection of these features is guided by their relevance to the classification task at hand.

#### Renaming Columns for Consistency
To ensure consistency in naming conventions and facilitate easier data manipulation and analysis, we rename the columns of the DataFrame. Specifically, we rename the 'File' column to 'file' to match the case and format of the other column names.

```python
# load annotated data
labelled_df = pd.read_csv('final_classificaiton_features.csv')
labelled_df = labelled_df[['File', 'PitchMean', 'PitchStd', 'SpectralCentroid', 'SpectralBandwidth', 'SpectralRolloff', 'Harmonic',
       'Percussive', 'Clarity', 'Confidence', 'Agreeableness', 'Pace',
       'Volume Dynamics', 'Compelling']]
labelled_df = labelled_df.rename(columns={'File':'file'})
labelled_df.columns
```

### Part 3: Merge Audio Features with Annotated Values to create Machine Learning Data Set

```python
# merge audio features with corresponding annotated label into new data set;
windowed_labels = extracted_df.merge(labelled_df[['file','Agreeableness', 'Compelling']], how='left', on=['file'], validate='many_to_one')
windowed_labels.set_index(keys=['file', 'file_id'],inplace=True)
data = windowed_labels.dropna()
data
```

### Part 4: Merge Aggregated Audio Features with Annotated Values to create Visualization Data Set

```python
merged_df = avg_extracted_df.merge(labelled_df[['file','Agreeableness', 'Compelling']], how='inner', on=['file'], validate='one_to_one')
merged_df.set_index(keys=['file'],inplace=True)
```

## Data Preprocessing

```python
# process Machine Learning Data (segmented data) into X, y for machine learning
X_data = data.drop(['Agreeableness','Compelling'], axis=1).to_numpy()
y_data = data[['Agreeableness','Compelling']].to_numpy()

# get train/validate/test datasets at a split of 60/20/20
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)
```

```python
# process aggregated data into X, y for visualization
X_agg_data = merged_df.drop(['Agreeableness','Compelling'], axis=1).to_numpy()
y_agg_data = merged_df[['Agreeableness','Compelling']].to_numpy()
```

```python
# --- obtain 2 principal components
pca = PCA(n_components=6)

principalComponents = pca.fit_transform(X_agg_data)
# print(f"principal components: {principalComponents} \n")
print("pca.explained_variance_ratio:", pca.explained_variance_ratio_, "\n")

pca_df = pd.DataFrame(data=principalComponents)

principalComponents = pca_df.drop([2,3,4,5], axis=1).to_numpy()
# principalComponents
```

```python
pca_df = pd.DataFrame(data=principalComponents)
pca_df
```

```python
# --- scree plot to visualize new features and the amount of variance it explains in the data
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'ro-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.show()
```

### K Means Clustering

```python
from sklearn.cluster import KMeans
# determine optimal value of k for K-Means Clustering

k_vals = [1, 2, 4, 8, 16]
k_cost = { k: [] for k in k_vals }
trials_per_k = 30

for k in k_vals:
    for trial in range(trials_per_k):
        KM = KMeans(n_clusters = k, max_iter = 500)
        KM.fit(principalComponents)
         
        # calculates squared error
        # for the clustered points
        k_cost[k].append(KM.inertia_)

# Might want to turn this into a simple list later for the silhouette plot
k_cost_means = dict(map(lambda kv: (kv[0], np.mean(kv[1])), k_cost.items()))

# plot the cost against K values
plt.plot(k_vals, k_cost_means.values(), color ='g', linewidth ='3')
plt.xlabel("Value of K")
plt.ylabel("Squared Error (Cost)")
plt.show() # clear the plot
```

```python
def show_clusters(data):
    # code adapted from https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html#sphx-glr-auto-examples-cluster-plot-kmeans-silhouette-analysis-py
    """ shows color-coded clusters from K means for various cluster numbers"""
    range_n_clusters = list(range(2,12,2))

    for n_clusters in range_n_clusters:
        # Create a subplot with 1 row and 2 columns
        fig, (ax1, ax2) = plt.subplots(1, 2)
        fig.set_size_inches(9, 4)

        # The 1st subplot is the silhouette plot
        # The silhouette coefficient can range from -1, 1 but in this example all
        # lie within [-0.1, 1]
        ax1.set_xlim([-0.1, 1])
        # The (n_clusters+1)*10 is for inserting blank space between silhouette
        # plots of individual clusters, to demarcate them clearly.
        ax1.set_ylim([0, len(data) + (n_clusters + 1) * 10])

        # Initialize the clusterer with n_clusters value and a random generator
        # seed of 10 for reproducibility.
        clusterer = KMeans(n_clusters=n_clusters, random_state=0)
        cluster_labels = clusterer.fit_predict(data)

        # The silhouette_score gives the average value for all the samples.
        # This gives a perspective into the density and separation of the formed
        # clusters
        silhouette_avg = silhouette_score(data, cluster_labels)
        print(
            "For n_clusters =",
            n_clusters,
            "The average silhouette_score is :",
            silhouette_avg,
        )

        # Compute the silhouette scores for each sample
        sample_silhouette_values = silhouette_samples(data, cluster_labels)

        y_lower = 10
        for i in range(n_clusters):
            # Aggregate the silhouette scores for samples belonging to
            # cluster i, and sort them
            ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

            ith_cluster_silhouette_values.sort()

            size_cluster_i = ith_cluster_silhouette_values.shape[0]
            y_upper = y_lower + size_cluster_i

            color = cm.nipy_spectral(float(i) / n_clusters)
            ax1.fill_betweenx(
                np.arange(y_lower, y_upper),
                0,
                ith_cluster_silhouette_values,
                facecolor=color,
                edgecolor=color,
                alpha=0.7,
            )

            # Label the silhouette plots with their cluster numbers at the middle
            ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

            # Compute the new y_lower for next plot
            y_lower = y_upper + 10  # 10 for the 0 samples

        ax1.set_title("The silhouette plot for the various clusters.")
        ax1.set_xlabel("The silhouette coefficient values")
        ax1.set_ylabel("Cluster label")

        # The vertical line for average silhouette score of all the values
        ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

        ax1.set_yticks([])  # Clear the yaxis labels / ticks
        ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

        # 2nd Plot showing the actual clusters formed
        colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
        ax2.scatter(
            data[:, 0], data[:, 1], marker=".", c=colors, lw=0, alpha=0.7,  edgecolor="k"
        )

        # Labeling the clusters
        centers = clusterer.cluster_centers_
        # Draw white circles at cluster centers
        ax2.scatter(
            centers[:, 0],
            centers[:, 1],
            marker="o",
            c="white",
            alpha=1,
            # s=200,
            edgecolor="k",
        )

        for i, c in enumerate(centers):
            ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=0.3, edgecolor="k")

        ax2.set_title("The visualization of the clustered data.")
        ax2.set_xlabel("Feature space for the 1st feature")
        ax2.set_ylabel("Feature space for the 2nd feature")
        # ax2.set_xlim([-60, -45])

        plt.suptitle(
            "Silhouette analysis for KMeans clustering on sample data with n_clusters = %d"
            % n_clusters,
            # fontsize=14,
            fontweight="bold",
        )

    plt.show()
```

## multi-output regression

```python
nn = Sequential()
nn.add(Dense(32, input_dim=X_data.shape[1], activation='relu'))
nn.add(Dense(64, activation='relu'))
nn.add(Dropout(rate=0.2))
nn.add(Dense(y_data.shape[1],activation='linear')) # 2 output labels; agreeableness & compelling
nn.compile(optimizer='adam', loss='mse', metrics=['accuracy'])

nn.summary()
```

```python
history = nn.fit(X_train, y_train, epochs=10, batch_size=1048, validation_data=(X_val, y_val))
```

```python
y_pred = nn.predict(X_test) 
# y_pred
```

## Model Performance

```python
print("Model Results for Agreeableness:")
mae = np.mean(np.abs(y_pred[:,0] - y_test[:,0]))
print("Mean Absolute Error:", mae)

# Calculate mean squared error (MSE)
mse = np.mean((y_pred[:,0] - y_test[:,0]) ** 2)
print("Mean Squared Error:", mse)

print ('-------------------------------------')
print("Model Results for Compelling:")

mae = np.mean(np.abs(y_pred[:,1] - y_test[:,1]))
print("Mean Absolute Error:", mae)

# Calculate mean squared error (MSE)
mse = np.mean((y_pred[:,1] - y_test[:,1]) ** 2)
print("Mean Squared Error:", mse)
```
